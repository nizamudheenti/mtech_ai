> ### Implement gradient descent using a learning rate of  = 0.07. Initialize the parameters to  = 0 (i.e., 0 = 1 = 0), and run one iteration of gradient descent from this initial starting point. Display the value of 0 and 1 that you get after the first iteration. 

```python
theta_0=0
theta_1=0
alpha=0.07 #learning rate
m = float(len(y)) #number of samples
y_current = (theta_0 * x) + theta_1  
cost = np.sum([data**2 for data in (y-y_current)]) / 2*m 
theta_0_gradient = -(1/m) * np.sum(x * (y - y_current)) 
theta_1_gradient = -(1/m) * np.sum(y - y_current)
theta_0 = theta_0 - (alpha * theta_0_gradient)
theta_1 = theta_1 - (alpha * theta_1_gradient)
print(theta_0, theta_1, cost)
```

> ### Continue running gradient descent for more iterations until  converges. After convergence, display the final values of 0 and 1 that you get, and plot the straight line fit from your algorithm on the same graph as your training data according to .

```python
""" 
    defining linear regression for univariate problem
    x=feature
    y=target
    theta_0 initial value 0
    theta_1 initial value 0
    epochs=number of iterations
"""
def lr(x, y, theta_0=0, theta_1=0, epochs=1, alpha=0.07):
    m = float(len(y))
    cost_array=[]
    diff=[]
    for i in range(epochs):
        y_current = (theta_0 * x) + theta_1
        cost = np.sum([data**2 for data in (y-y_current)]) / 2*m
        theta_0_gradient = -(1/m) * np.sum(x * (y - y_current))
        theta_1_gradient = -(1/m) * np.sum(y - y_current)
        theta_0 = theta_0 - (alpha * theta_0_gradient)
        theta_1 = theta_1 - (alpha * theta_1_gradient)
        cost_array.append(cost)
        if(i>0):
            diff.append(cost_array[i-1]-cost_array[i])
    return theta_0, theta_1, cost,cost_array,diff
```

```python
epoch=100
theta_0,theta_1,cost,cost_array,diff=lr(x,y,epochs=epoch)
```

#### reading data from file

```python
x_file=open('2-D_x.dat') #opening file containing feature/x data
y_file=open('2-D_y.dat') #opening file containing target/y data
x1=[]#initializing x array for storing
x2=[]
y=[] #initializing y array for storing
for i in x_file: #iterating each element from x_file
    x1.append(float(i.split()[0]))#converting str to float and appending to x1 data list
    x2.append(float(i.split()[1]))#converting str to float and appending to x2 data list
for j in y_file:#iterating each element from y_file
    y.append(float(j))#converting str to float and appending to y data list
    
```

#### creating dataframe for storing data

```python
data=pd.DataFrame()
```

```python
data['living']=x1
data['bed']=x2
data['price']=y
```

```python
data.head()
```

#### data normalization 

```python
data = (data - data.mean()) / data.std()
data.head()
```

#### cost function

```python
def J_theta(X, y, theta):
    inner = np.power(((X * theta.T) - y), 2)
    return np.sum(inner) / (2 * len(X))
```

#### Gradient Descent 

```python
def GD(X, y, theta, alpha, iters):
    temp = np.zeros(theta.shape)
    parameters = X.shape[1]
    cost = np.zeros(iters)
    for i in range(iters):
        error = (X * theta.T) - y
        for j in range(parameters):
            term = np.multiply(error, X[:,j])
            temp[0,j] = theta[0,j] - (alpha*(1 /len(X)) * np.sum(term))    
        theta = temp
        cost[i] = J_theta(X, y, theta)  
    return theta, cost
```

```python
data.insert(0, 'Ones', 1)
# set X (training data) and y (target variable)
X = data.iloc[:,0:3]
y = data.iloc[:,3:4]
# convert to matrices and initialize theta
X = np.matrix(X.values)
y = np.matrix(y.values)
theta = np.matrix(np.array([0,0,0]))
```

```python
weight =[] #array for storing weights
j=[] #array for storing cost
alpha=[0.01,0.05,0.1,0.5,0.7,1] # learning rates/alpha
iters=50 #epochs
for i in alpha:
    w, cost = GD(X, y, theta,i, iters)
    weight.append(w) #appending theta values
    j.append(cost)  #appending cost values
for i in range(len(alpha)):
    plt.plot(range(0,iters), j[i],label="alpha "+str(alpha[i]))#ploting cost vs epoch for each alpha values
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('cost vs. Training Epoch')
plt.legend()
plt.show()
```
